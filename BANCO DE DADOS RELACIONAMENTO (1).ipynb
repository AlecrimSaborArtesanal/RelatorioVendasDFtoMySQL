```python

# Defina o caminho da pasta contendo os arquivos CSV
caminho_pasta = r'C:\Users\kimge\OneDrive\Power BI\MODELAGEM_ALECRIM'

# Crie um dicionário vazio para armazenar os dataframes
dataframes = {}

# Iterar sobre os arquivos na pasta
for arquivo in os.listdir(caminho_pasta):

    # Verifique se o arquivo é um CSV
    if arquivo.endswith('.csv'):

        # Obtenha o nome do arquivo sem a extensão .csv
        nome_arquivo = os.path.splitext(arquivo)[0]

        # Crie o caminho completo para o arquivo
        caminho_arquivo = os.path.join(caminho_pasta, arquivo)

        # Leia o arquivo CSV em um dataframe e adicione-o ao dicionário de dataframes
        dataframes[nome_arquivo] = pd.read_csv(caminho_arquivo)

        # Alternativamente, se você quiser salvar o dataframe como um arquivo CSV com o mesmo nome do arquivo original:
        # dataframes[nome_arquivo].to_csv(os.path.join(caminho_pasta, nome_arquivo + '.csv'), index=False)

# Agora você pode acessar cada dataframe pelo nome do arquivo correspondente, por exemplo:
print('As primeiras 5 linhas do dataframe correspondente ao arquivo "CARDAPIO":')
print(dataframes['CARDAPIO'].head())

for nome_arquivo, df in dataframes.items():
    globals()[nome_arquivo] = df.copy()
```

```python
CREATE TABLE vendas (
    data DATE NOT NULL,
    forma_pagamento VARCHAR(20),
    hora TIME NOT NULL,
    id_cliente INT,
    nome_cliente VARCHAR(50),
    id_produto INT NOT NULL,
    id_venda INT NOT NULL,
    nome_produto VARCHAR(50) NOT NULL,
    valor_unitario DECIMAL(10, 2) NOT NULL,
    quantidade INT NOT NULL,
    total_venda DECIMAL(10, 2) NOT NULL,
    PRIMARY KEY (id_venda)
);

CREATE TABLE cardapio (
    id_produto INT NOT NULL,
    nome_produto VARCHAR(50) NOT NULL,
    preco_venda DECIMAL(10, 2) NOT NULL,
    categoria VARCHAR(20) NOT NULL,
    PRIMARY KEY (id_produto)
);

CREATE TABLE estoque (
    data_entrada DATE NOT NULL,
    id_item INT NOT NULL,
    nome_item VARCHAR(50) NOT NULL,
    quantidade_estoque INT NOT NULL,
    unidade VARCHAR(10) NOT NULL,
    valor_unitario DECIMAL(10, 2) NOT NULL,
    PRIMARY KEY (data_entrada, id_item)
);

CREATE TABLE compras (
    data_compra DATE NOT NULL,
    forma_pagamento VARCHAR(20) NOT NULL,
    fornecedor VARCHAR(50) NOT NULL,
    id_compra INT NOT NULL,
    id_item INT NOT NULL,
    nome_item VARCHAR(50) NOT NULL,
    quantidade INT NOT NULL,
    unidade VARCHAR(10) NOT NULL,
    valor_unit DECIMAL(10, 2) NOT NULL,
    PRIMARY KEY (id_compra, id_item)
);

CREATE TABLE ficha_tecnica (
    categoria VARCHAR(20) NOT NULL,
    id_produto INT NOT NULL,
    nome_produto VARCHAR(50) NOT NULL,
    nome_item VARCHAR(50) NOT NULL,
    quantidade DECIMAL(10, 2) NOT NULL,
    unidade VARCHAR(10) NOT NULL,
    tempo_preparo INT NOT NULL,
    valor_unitario DECIMAL(10, 2) NOT NULL,
    custo_total DECIMAL(10, 2) NOT NULL,
    descricao TEXT,
    PRIMARY KEY (categoria, id_produto, nome_item)
);
```

```python
import mysql.connector

# cria conexão com o banco de dados
cnx = mysql.connector.connect(
        host = 'localhost',
        user = 'root',
        password = 'alecrim',
        database = 'alecrim',
)

# cria tabela clientes
cursor = cnx.cursor()
query = '''
    CREATE TABLE clientes (
        id_cliente INT AUTO_INCREMENT PRIMARY KEY,
        nome_cliente VARCHAR(50) NOT NULL,
        sexo enum('M', 'F'),
        nascimento date,
        nacionalidade VARCHAR(20) DEFAULT 'Brasil',
        email VARCHAR(50) UNIQUE,
        telefone VARCHAR(15)
    )
'''
cursor.execute(query)

# cria tabela cardapio
query = '''
    CREATE TABLE cardapio (
        id_produto INT AUTO_INCREMENT PRIMARY KEY,
        nome_produto VARCHAR(50) NOT NULL,
        descricao_prato VARCHAR(200) NOT NULL,
        preco_venda DECIMAL(10,2) NOT NULL,
        categoria VARCHAR(50) NOT NULL
    )
'''
cursor.execute(query)

# cria tabela ficha_tecnica
query = '''
    CREATE TABLE ficha_tecnica (
        id_item INT AUTO_INCREMENT PRIMARY KEY,
        id_produto INT NOT NULL,
        nome_produto VARCHAR(50) NOT NULL,
        nome_item VARCHAR(50) NOT NULL,
        descricao VARCHAR(200) NOT NULL,
        quantidade INT NOT NULL,
        unidade VARCHAR(10) NOT NULL,
        valor_unitario DECIMAL(10,2) NOT NULL,
        custo_total DECIMAL(10,2) NOT NULL,
        tempo_preparo TIME NOT NULL,
        categoria VARCHAR(50) NOT NULL,
        FOREIGN KEY (id_produto) REFERENCES cardapio(id_produto)
    )
'''
cursor.execute(query)

# cria tabela estoque
query = '''
    CREATE TABLE estoque (
        id_item INT AUTO_INCREMENT PRIMARY KEY,
        nome_item VARCHAR(50) NOT NULL,
        quantidade_estoque INT NOT NULL,
        unidade VARCHAR(10) NOT NULL,
        valor_unitario DECIMAL(10,2) NOT NULL,
    )
'''
cursor.execute(query)

# cria tabela compras
query = '''
    CREATE TABLE compras (
        id_compra INT AUTO_INCREMENT PRIMARY KEY,
        data_compra DATE NOT NULL,
        fornecedor VARCHAR(50) NOT NULL
    )
'''
cursor.execute(query)

# cria tabela vendas
query = '''
    CREATE TABLE vendas (
        id_venda INT AUTO_INCREMENT PRIMARY KEY,
        id_cliente INT,
        nome_cliente VARCHAR(50),
        id_produto INT NOT NULL,
        quantidade INT NOT NULL,
        valor_unitario DECIMAL(10,2) NOT NULL,
        total_venda DECIMAL(10,2) NOT NULL,
        forma_pagamento VARCHAR(50) NOT NULL,
        hora TIME NOT NULL,
        data DATE NOT NULL,
        FOREIGN KEY (id_cliente) REFERENCES clientes(id_cliente),
        FOREIGN KEY (id_produto) REFERENCES cardapio(id_produto)
    )
'''
cursor.execute(query)


# cria tabela compras_itens
query = '''
    CREATE TABLE compras_itens (
        id_compra INT NOT NULL,
        id_item INT NOT NULL,
        nome_item,
        quantidade INT,
        unidade VARCHAR(10) NOT NULL,
        preco_unitario DECIMAL(10,2),
        FOREIGN KEY (id_compra) REFERENCES compras(id_compra),
        FOREIGN KEY (id_item) REFERENCES estoque(id_item),
        PRIMARY KEY (id_compra, id_item)
    )
'''
cursor.execute(query)


# fecha conexão
cnx.close()


```
